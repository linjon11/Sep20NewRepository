```python
# let's import the survey data
# we'll skip the first row, and have to specify the text encoding
df = pd.read_csv('HIS_survey_data_analysis_clean.csv', encoding= 'unicode_escape', skiprows=[0])
df.head()
```

```python
# replace all missing responses with np.nan
df.fillna(np.nan, inplace = True)
```

```python
def plot_column(df, colname, newdfcols):
    ''' Function for plotting categorical columns
    inputs: df - name of dataframe, colname - name of column from dataframe,
            newdfcols - names of columns in dataframe that is transformed
    outputs: plot of all responses in df.colname as percentages
    '''
    # create a new dataframe that counts the number of values of each type of response
    # in the specified column
    newdf = pd.DataFrame(df[colname].value_counts()).reset_index()
    # rename the columns in the new dataframe
    newdf.columns = newdfcols
    # find the number of individuals who didn't answer this question
    newdf.loc[len(df.index)] = ['No Answer', df[colname].isnull().sum()]
    # find percentages for all responses to question
    newdf[newdfcols[1]] = newdf[newdfcols[1]]/len(df)*100
    
    # make plot
    sns.set_style("whitegrid")
    sns.barplot(data=newdf, x=newdf.columns[0], y=newdf.columns[1])
    plt.xticks(rotation = 90)
    plt.show()
```
